 <h3> Part One : Fetching Data  </h3>

```python
"""In this block we have created a function to read and absorb the required fields out of my pay slips"""

# importing necessary packages
from tabula import read_pdf
import pandas as pd

# this is an empty data frame to which all necessary data will ne appended 
intermideate_df = pd.DataFrame( index =['Basic Salary', 
                 'Conveyance Non Taxable', 
                 'House Rent Allowance',
                 'Sundry Med￼
ical',
                 'Leave Travel Allowance',
                 'Personal Allowance',
                 'City Allowance',
                 'BoB Kitty Allowance',
                 'Miscellaneous',
                 'Performance Pay',
                 'Total Earnings  (Current + Arrears)' ])


# this is a cute function to read pdf and synthesis necessary fields out of pdf, finally appending data to df
def read_data(file_name):
   
    path   = "/home/jasp/Documents/pay_slips/"
    df_version1 = read_pdf(path+file_name+".pdf")
    df_version1 = df_version1.rename(columns = {"Mr. JASPREET  SINGH":"Associate","Unnamed: 2": file_name})
    df_version2 = df_version1.loc[: , ["Associate", file_name]]
    df_version3 = pd.DataFrame(df_version2)
    df_version4 = df_version3.set_index('Associate')
    df_version5 = df_version4.loc[['Basic Salary', 
                 'Conveyance Non Taxable', 
                 'House Rent Allowance',
                 'Sundry Medical',
                 'Leave Travel Allowance',
                 'Personal Allowance',
                 'City Allowance',
                 'BoB Kitty Allowance',
                 'Miscellaneous',
                 'Performance Pay',
                 'Total Earnings  (Current + Arrears)' 
                 ]];
    
    global intermideate_df
    
    intermideate_df = pd.concat([ intermideate_df, df_version5], axis=1)
    
    return intermideate_df;
 
```

```python
# this is a sample call to read data function , which collects data 
read_data("July_2018")
```

```python
# here we have converted the cumulative data frame to csv for future reference 
intermideate_df.to_csv(r'/home/jasp/Documents/pay_slips/Aug_2018.csv')
```
